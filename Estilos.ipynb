```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
#Lista de estilos para las graficas
print(plt.style.available)
```

```python
#Para ver un ejemplo visual de todos los estilos de grafica
for i in plt.style.available:
    plt.style.use(i)
    print(i)
    fig, ax = plt.subplots(figsize=(5,5))
    ax.plot(x)
    plt.show()
```

```python
plt.style.use('fivethirtyeight')
```

## Estilo Matlab

```python
plt.style.use("dark_background")
fig,ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'g-')
ax.plot(x,x+3,'b.-')
ax.plot(x,x+4,'y:')
```

## Estilo Pyplot

```python
plt.style.use("dark_background")
fig,ax = plt.subplots(figsize=(8,8))
#alpha = nivel de transparencia
#linewidth = grosor de linea
ax.plot(x,x+1,color='green', alpha=0.5, linewidth = 5, linestyle='-', marker='x', markersize=10)
ax.plot(x,x+2,color='blue', linewidth = 10, linestyle='--')
ax.plot(x,x+3,color='#66FF66', linewidth = 2, linestyle='dashed', marker='v')
ax.plot(x,x+4,color='#CCFFCC', linewidth = 0.9, linestyle=':', marker='P')
```
