```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#carga un dataset de ejemplo, de propinas en un restaurante
tips = sns.load_dataset('tips')

tips
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist',
            marginal_ticks=True,marginal_kws=dict(bins=25,fill=False,multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex',palette='coolwarm',corner=True)
plt.show()
```
