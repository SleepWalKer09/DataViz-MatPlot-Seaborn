```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

# Histograma

```python
#bins = numero de barras para indicar el rango de frecuencia de datos
plt.hist(data, bins=20,histtype='step')
plt.show()
```

# Boxplot

```python
data = np.append(data,200)
#showfliers = visualizar datos outliers
plt.boxplot(data,vert=False, patch_artist=True,notch=True, showfliers=False)
plt.show()
```

# Scatter

```python
N = 50
x= np.random.rand(N)
y= np.random.rand(N)

area = (30* np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area, c=colors,marker='o',alpha= 0.5)
plt.show()
```
